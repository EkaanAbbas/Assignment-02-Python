```python
# c1 centroid_point1
# c2 centroid_point2
# c3 centroid_point3
# c4 centroid_point4
#sum1 (It takes the sum of first group elements)
#sum2 (It takes the sum of second group elements)
#no_of_counts_01 (It will count the value of group 01 and it will help to find the average of group 01 for the next centroid_point)
#no_of_counts_02 (It will count the value of group 02 and it will help to find the average of group 02 for the next centroid_point)

list1 = [15,15,16,19,19,20,20,21,22,28,35,40,41,42,43,44,60,61,65]
    
print("List is", list1)

centroid_point1 = 16
centroid_point2 = 22
centroid_point3 = 0
centroid_point4 = 0

count1 = 0

while count1 == 0:
    
    sum1 = 0
    sum2 = 0
    no_of_count_01 = 0
    no_of_count_02 = 0
    group1 = []
    group2 = []
    average = 0
    average2 = 0
    
    for i in range(len(list1)):
        
        x = int(list1[i]) - centroid_point1
        x = abs(x)
     
        y = int(list1[i]) - centroid_point2
        y = abs(y)
            
        if(x > y):
            group2.append(list1[i])
            print("Group2", group2)
            no_of_count_02 = no_of_count_02 + 1 
            
        elif (x < y):
            group1.append(list1[i])
            print("Group1", group1)
            no_of_count_01 = no_of_count_01 + 1
            
        else:
            no_of_count_02 = no_of_count_02 + 1
            group2.append(list1[i])
            print("Group2", group2)

    if(no_of_count_01 != 0):
        for i in range(no_of_count_01):
            sum1 = sum1 + group1[i]
            
        average = sum1 / no_of_count_01
        centroid_point3 = average
        print("average", average)
        
    else:
        centroid_point3 = centroid_point1
    
    if(no_of_count_02 != 0):
        for i in range(no_of_count_02):
            sum2 = sum2 + group2[i]
            
        average2 = sum2 / no_of_count_02
        centroid_point4 = average2
        print("average2", average2)
        
    else:
        centroid_point4 = centroid_point2
        
    if(centroid_point1 == centroid_point3 and centroid_point2 == centroid_point4):
        count1 = count1 + 1
    else:
        centroid_point1 = centroid_point3
        centroid_point2 = centroid_point4
    
print("c1",centroid_point1)
print("c2",centroid_point2)
print("c3", centroid_point3)
print("c4", centroid_point4)
```
